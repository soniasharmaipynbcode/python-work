Data Types in Python 

Built-in Data Types
In programming, data type is an important concept.

Variables can store data of different types, and different types can do different things.

```python
print('diffrent methods for write data types')
list=[1,2,3,4,5,6]
set={7,8,9,10}
tuple=('a','b','c','d')
dictionary={'name':'sumit','grade':2}
```

Display diffrent data types with examples

```python
print('strint data type')
x = "Hello World"

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('interger data type')
x = 20

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('float data type')
x = 20.5

#display x:
print(x)

#display the data type of x:
print(type(x)) 
x = 1j

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('list data type')
x = ["apple", "banana", "cherry"]

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('tuple data type')
x = ("apple", "banana", "cherry")

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('range')
x = range(6)

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('dictionary')

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('set data type')
x = {"apple", "banana", "cherry"}

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('set frozen')
x = frozenset({"apple", "banana", "cherry"})

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('boolean')
x = True

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('byte')
x = b"Hello"

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('byte array')
x = bytearray(5)

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('memory view')
x = memoryview(bytes(5))

#display x:
print(x)

#display the data type of x:
print(type(x)) 
print('none data type')
x = None

#display x:
print(x)

#display the data type of x:
print(type(x)) 
```
